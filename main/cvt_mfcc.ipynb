```python
import numpy as np
import os
import cv2
import pandas as pd
import sklearn
import librosa
import shutil
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings(action='ignore')
```

```python
train_path = './train/'
test_path = './test/'
```

```python
train_df = pd.read_csv('./train.csv')
test_df = pd.read_csv('./test.csv')
```

```python
train_wav = os.listdir(train_path)
train_wav
```

```python
for wav in train_wav:
    audio, sr = librosa.load(os.path.join(train_path, wav), sr=22050)
    mfcc = librosa.feature.mfcc(audio, sr, n_mfcc=100, n_fft=552, hop_length=221)
    mfcc = sklearn.preprocessing.scale(mfcc, axis=1)
    
    
    plt.figure(figsize=(10, 4))
    plt.imshow(mfcc, aspect='auto', origin='lower', cmap='viridis')
    plt.tight_layout()

    fn = os.path.join('./train_mfcc/', str(wav[:-4]) + '.jpeg')
    plt.savefig(fn, bbox_inches='tight')
    plt.close()
```

```python
test_wav = os.listdir(test_path)
test_wav
```

```python
for wav in test_wav:
    audio, sr = librosa.load(os.path.join(test_path, wav), sr=22050)
    mfcc = librosa.feature.mfcc(audio, sr, n_mfcc=100, n_fft=552, hop_length=221)
    mfcc = sklearn.preprocessing.scale(mfcc, axis=1)
    
    plt.figure(figsize=(10, 4))
    plt.imshow(mfcc, aspect='auto', origin='lower', cmap='viridis')
    plt.tight_layout()
    
    fn = os.path.join('./test_mfcc/', str(wav[:-4]) + '.jpeg')
    plt.savefig(fn, bbox_inches='tight')
    plt.close()
```
